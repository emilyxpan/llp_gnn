```python
import uproot
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep
import hist
```

```python
fileset = {} #dictionary of lists of directories

sig_dir = '/ceph/cms/store/user/aaportel/B-Parking/rechits_v2/BToKPhi_MuonLLPDecayGenFilter_PhiToPi0Pi0_mPhi0p3_ctau300/'
fileset['sample'] = [sig_dir + f'BToKPhi_MuonLLPDecayGenFilter_PhiToPi0Pi0_mPhi0p3_ctau300_{str(i).zfill(7)}.root' for i in range(328)]

f = uproot.open(fileset['sample'][0]) #first path in dictionary
raw_data = f['MuonSystem'].arrays(library = 'ak')
```

```python
raw_data.fields
```

```python
cscRechitsx = ak.flatten(raw_data.cscRechitsX)
cscRechitsy = ak.flatten(raw_data.cscRechitsY)
cscRechitsr = np.sqrt((cscRechitsx**2 + cscRechitsy**2))
```

```python
h = hist.Hist.new.Regular(50, 0, 1100, name="cscRechits r [cm]").Double()
h.fill(cscRechitsr)
hep.histplot(h)
```

```python
h = hist.Hist.new.Regular(50, -np.pi, np.pi, name="cscRechits $\phi$").Double()
h.fill(ak.flatten(raw_data.cscRechitsPhi))
hep.histplot(h)
```

```python
h = hist.Hist.new.Regular(50, -4, 4, name="cscRechits $\eta$").Double()
h.fill(ak.flatten(raw_data.cscRechitsEta))
hep.histplot(h)
```

```python
h = hist.Hist.new.Regular(20, -1100, 1100, name="cscRechits Z [cm]").Double()
h.fill(ak.flatten(raw_data.cscRechitsZ))
hep.histplot(h)
```

```python
h = hist.Hist.new.Regular(50, 0, 1100, name="gLLP r [cm]").Double()
h.fill(raw_data.gLLP_decay_vertex_r)
hep.histplot(h)
```
